## Importing the Dataset

```python
import pandas as pd
import numpy as np
```

## Reading the Dataset

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CS 430 Machine Learning/XGBoost_Presentation/customer_data.csv")
```

## Encoding Categorical Data in the Dataset

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
data['fea_2'] = imputer.fit_transform(data[['fea_2']])
```

## Splitting the Data into X and Y

```python
x = data.drop('label', axis = 1)
y = data['label']
```

## Training, Testing, and Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0, stratify = y)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train, y_train)
```

```python
dt_model.score(x_test, y_test)
```

```python
dt_model.score(x_train, y_train)
```

## XGBoost Classifier

```python
import xgboost as xgb
```

```python
xgb_classifier = xgb.XGBClassifier()
```

```python
xgb_classifier.fit(x_train, y_train)
```

```python
xgb_classifier.score(x_test, y_test)
```

```python
xgb_classifier.score(x_train, y_train)
```

```python
y_pred = xgb_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
